### (Unsupervised) Principle Component Analysis (PCA)
![Co-Variance.png](https://raw.githubusercontent.com/vaibhaVishwakarma/ML-from-scratch/master/NoteBooks/Co-Variance.png)

```python
import numpy as np
class PCA:
    def __init__(self , n_components):
        self.comonents = None
        self.n_components = n_components

    def fit(self , X):
        """
        eigen values correspond to the varience of the daya so higher the varience along the vector higher the eigen value
        """
        cov_matrix = np.cov(X.T) # transposed to consider features as records and cov is calculated for records only

        eig_vecs , eig_vals = np.linalg.eig( cov_matrix )
        eig_vecs = eig_vecs.T # as vectors are represented as columns -> now as rows

        idxs = np.argsort(eig_vals)

        self.components = eig_vecs[idxs][:self.n_components]

    def transform(self , X):
        return np.dot(X , self.components.T) # m x n dot (n_components x n).T -> m x n_components
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
data = datasets.load_iris()
X , y = data.data , data.target
X_train , X_test , y_train , y_test = train_test_split( X , y ,
                                                       train_size= 0.2 , 
                                                       shuffle= True,
                                                       random_state=4444)
```

```python
pca = PCA(3)
pca.fit(X_train)
transformed_test = pca.transform(X_test)
transformed_test.shape
```

```python
np.vstack([transformed_test.T,y_test]).T.shape
```

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame(np.vstack([transformed_test.T,y_test]).T , columns = ['x','y','z','class'])

fig = px.scatter_3d(df, color='class' , x = 'x' , y = 'y' , z = 'z' )
fig.show()
```

```python
pca = PCA(2)
pca.fit(X_train)
transformed_test = pca.transform(X_test)

df = pd.DataFrame(np.vstack([transformed_test.T , y_test]).T , columns = ['x' , 'y' , 'class'])

fig = px.scatter(df , x = 'x' , y ='y' , color = 'class' )
fig.show()
```
